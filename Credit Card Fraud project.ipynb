```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
# data information
df.info()
```

```python
# lets check if there are any missing values
df.isnull().sum()
```

```python
# lets check the distribution of normal transactions and fraud transactions
df['Class'].value_counts()
```

```python
# it is clear that this dataset is very unbalanced
# 0 represents normal transactions and 1 represents a fraudulent transaction
# we have to seperate the data for analysis

legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
# statistics for the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions

df.groupby('Class').mean()
```

```python
# we will use undersampling
# we will build a sample dataset containing a similar distribution of normal transaction and fraud transactions

legit_sample = legit.sample(n=492)
```

```python
# Now we concatinate the two dataframes

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# now we split the dataset into features and targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# now we split the data into training and testing set.
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Now we train the model using Logistic regression

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Now lets evaluate the model
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```
